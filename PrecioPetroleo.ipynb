```python
#Análisis del Precio de Petroleo WTI en los últimos 10 años
#La Data incluye un reporte mensual del precio de petroleo WTI desde el 4-Marzo-2013 hasta el 4-Marzo-2023


#Librerias
import pandas as pd
import matplotlib.pyplot as plt
from pandas import DataFrame

#Importar Data
data_petroleo = pd.read_excel('/content/Precio_WTI_MENSUAL_P.xlsx')
data_petroleo
df=DataFrame(data_petroleo)
df
```

```python
#Tendencia
fig, ax=plt.subplots(figsize=(13,9))
ax.plot(data_petroleo["Fecha"], data_petroleo["Último"], marker="x", label='Últimos precios', color="tab:red")
ax.plot(data_petroleo["Fecha"], data_petroleo["Apertura"], color="tab:blue", marker="x", label='Precios de apertura')

ax.invert_xaxis()
ax.set_xlim([1,120])
ax.set_xticks(range(1,132,10))
ax.set_title("Tendencia de los precios del petróleo", loc="center",
             fontdict={"fontsize":15, "fontweight":"bold"})

ax.set_xlabel("Fecha", fontdict={"fontsize":15, "fontweight":"bold", "color":"tab:red"})
ax.set_ylabel("Precio", fontdict={"fontsize":15, "fontweight":"bold", "color":"tab:red"})

plt.legend(loc='lower left', fontsize="large")
df.plot(title="Diagrama de dispersión de precios de petróleo en los últimos 10 años", x='Máximo', y='Mínimo', kind='scatter', figsize=(13,9))
ax.set_xticks(range(1,132,12))

ax.grid() #mostrar cuadricula
plt.show
```

```python
fig, ax = plt.subplots(figsize=(14,10))
ax.bar(data_petroleo['Fecha'], data_petroleo['Último'])
ax.set_xlim([1,120])
ax.set_xticks(range(1,132,10))

ax.set_title("Diagrama de Barras Precios del Petróleo", loc="center",
             fontdict={"fontsize":15, "fontweight":"bold"})
ax.set_xlabel("Fecha", fontdict={"fontsize":15, "fontweight":"bold", "color":"tab:Blue"})
ax.set_ylabel("Precio", fontdict={"fontsize":15, "fontweight":"bold", "color":"tab:blue"})

ax.grid() #mostrar cuadricula
plt.show()
```
